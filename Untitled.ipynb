```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# Create a sample network (graph)
# -------------------------------
n = 10  # number of polygons/nodes
np.random.seed(42)

# Create random positions for the nodes (these represent the centroids of the polygons)
positions = {i: (np.random.uniform(0, 10), np.random.uniform(0, 10)) for i in range(n)}

# Initialize graph and add nodes with an initial phenomenon value Y (drawn at random)
G = nx.Graph()
for i in range(n):
    G.add_node(i, pos=positions[i], Y=np.random.rand())
```

```python
# Define adjacency based on a threshold distance (simulate shared boundaries)
adjacency_threshold = 3.0  # if centroids are closer than this, we consider polygons "adjacent"
for i in range(n):
    for j in range(i + 1, n):
        d = np.linalg.norm(np.array(positions[i]) - np.array(positions[j]))
        if d < adjacency_threshold:
            # Add an edge with weight = 1 (binary adjacency)
            G.add_edge(i, j, weight=1)
```

```python
# ---------------------------------------------
# Define functions that simulate phenomenon models
# ---------------------------------------------

def local_model(G):
    """
    Local (Within Polygon): The phenomenon Y is solely a function of internal factors.
    Here we simply return the current node attribute 'Y'.
    """
    return {i: G.nodes[i]['Y'] for i in G.nodes}

def subnational_model(G, beta0=0.5, beta1=0.3):
    """
    Subnational (Cluster of Neighbors):
    Y_i = beta0 + beta1 * sum_{j in N(i)} Y_j + noise.
    """
    new_Y = {}
    for i in G.nodes:
        neighbors = list(G.neighbors(i))
        sum_neighbors = sum(G.nodes[j]['Y'] for j in neighbors)
        noise = np.random.normal(scale=0.1)
        new_Y[i] = beta0 + beta1 * sum_neighbors + noise
    return new_Y

def national_model(G, weight_decay=0.1):
    """
    National/International (Neighbors + Non-Neighbors):
    Y_i = f( sum_j w_{ij} Y_j ), with weights w_{ij} that decay with distance.
    Here we use an exponential decay function for the weights.
    """
    new_Y = {}
    for i in G.nodes:
        xi = np.array(G.nodes[i]['pos'])
        weighted_sum = 0
        total_weight = 0
        for j in G.nodes:
            if i == j:
                continue
            xj = np.array(G.nodes[j]['pos'])
            d = np.linalg.norm(xi - xj)
            # Weight decays exponentially with distance
            w = np.exp(-weight_decay * d)
            weighted_sum += w * G.nodes[j]['Y']
            total_weight += w
        new_Y[i] = weighted_sum / total_weight if total_weight != 0 else G.nodes[i]['Y']
    return new_Y

def clear_boundary_model(G, state_assignment, gamma=0.7):
    """
    Local-Subnational (Clear Boundaries):
    Each node belongs to one state and
    Y_i = alpha + gamma * (state effect) + noise.
    """
    # Each state is assigned a random effect value.
    states = set(state_assignment.values())
    state_effects = {s: np.random.rand() for s in states}
    new_Y = {}
    alpha = 0.2
    for i in G.nodes:
        state = state_assignment[i]
        noise = np.random.normal(scale=0.1)
        new_Y[i] = alpha + gamma * state_effects[state] + noise
    return new_Y

def non_neighbor_model(G):
    """
    Local-National/International (Non-Neighbors):
    Consider only non-neighboring nodes (i.e., nodes without an edge) for the phenomenon.
    For each node, average the Y values of nodes that are not directly connected.
    """
    new_Y = {}
    for i in G.nodes:
        xi = np.array(G.nodes[i]['pos'])
        influence = 0
        count = 0
        for j in G.nodes:
            if i == j or G.has_edge(i, j):
                continue
            xj = np.array(G.nodes[j]['pos'])
            d = np.linalg.norm(xi - xj)
            # Optionally, use a condition on distance if needed.
            influence += G.nodes[j]['Y']
            count += 1
        new_Y[i] = influence / count if count > 0 else G.nodes[i]['Y']
    return new_Y

def fuzzy_membership_model(G, membership):
    """
    Local-Subnational (Fuzzy Boundaries):
    Each node has fuzzy membership in one or more regions.
    The overall effect is a weighted combination of region effects.
    
    'membership' is a dictionary:
        membership[node] = { region_id: degree_of_membership, ... }
    """
    new_Y = {}
    # Each region is given a random effect value.
    regions = {r for m in membership.values() for r in m.keys()}
    region_effects = {r: np.random.rand() for r in regions}
    
    for i in G.nodes:
        effect = sum(membership[i].get(r, 0) * region_effects[r] for r in membership[i])
        noise = np.random.normal(scale=0.1)
        new_Y[i] = effect + noise
    return new_Y
```

```python
# -------------------------
# Apply the various models
# -------------------------

# 1. Local model (node's own attribute)
Y_local = local_model(G)

# 2. Subnational model (neighbors influence)
Y_subnational = subnational_model(G)

# 3. National/International model (distanceâ€“weighted influence)
Y_national = national_model(G)

# 4. Clear boundary model: assign each node a state (for simplicity, randomly choose between 'A' and 'B')
states = ['A', 'B']
state_assignment = {i: np.random.choice(states) for i in G.nodes}
Y_clear_boundary = clear_boundary_model(G, state_assignment)

# 5. Local-National/International (non-neighbors only)
Y_non_neighbor = non_neighbor_model(G)

# 6. Fuzzy boundaries: assign each node a fuzzy membership in two regions ('R1' and 'R2')
membership = {}
for i in G.nodes:
    # Generate two random membership degrees that sum to 1.
    r1 = np.random.rand()
    r2 = np.random.rand()
    total = r1 + r2
    membership[i] = {'R1': r1 / total, 'R2': r2 / total}
Y_fuzzy = fuzzy_membership_model(G, membership)

# ------------------------------------
# Visualize one of the phenomenon models
# ------------------------------------
# For demonstration, we update the node attributes with the subnational model's Y values.
for i in G.nodes:
    G.nodes[i]['Y_model'] = Y_subnational[i]

# Get positions and prepare labels for plotting
pos = nx.get_node_attributes(G, 'pos')
labels = {i: f"{G.nodes[i]['Y_model']:.2f}" for i in G.nodes}

plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10)
nx.draw_networkx_labels(G, pos, labels=labels, font_color='black')
plt.title("Subnational Model: Node Values Influenced by Neighbors")
plt.show()
```
